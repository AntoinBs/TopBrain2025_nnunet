# 2 - Preprocessing Pre-training datas (TopCow)

```python
image_tr_folder = '../data/raw_pretraining/TopCoW2024_Data_Release/imagesTr'
label_tr_folder = '../data/raw_pretraining/TopCoW2024_Data_Release/cow_seg_labelsTr'

output_img_ct_dir = '../data/raw_pretraining/reorganized/imagesTr_topcow_ct'
output_lbl_ct_dir = '../data/raw_pretraining/reorganized/labelsTr_topcow_ct'
output_img_mr_dir = '../data/raw_pretraining/reorganized/imagesTr_topcow_mr'
output_lbl_mr_dir = '../data/raw_pretraining/reorganized/labelsTr_topcow_mr'

os.makedirs(output_img_ct_dir, exist_ok=True)
os.makedirs(output_lbl_ct_dir, exist_ok=True)
os.makedirs(output_img_mr_dir, exist_ok=True)
os.makedirs(output_lbl_mr_dir, exist_ok=True)

for filename in os.listdir(label_tr_folder):
    if filename.endswith('.nii.gz'):
        filepath = os.path.join(label_tr_folder, filename)
        modality = filename.split('_')[1]

        if modality == 'ct':
            shutil.copy(filepath, os.path.join(output_lbl_ct_dir, filename))

        elif modality == 'mr':
            shutil.copy(filepath, os.path.join(output_lbl_mr_dir, filename))

        else:
            print(f"ERROR : Unknown modality in file {filename} : {modality}")
for filename in os.listdir(image_tr_folder):
    if filename.endswith('.nii.gz'):
        filepath = os.path.join(image_tr_folder, filename)
        modality = filename.split('_')[1]
        new_filename = filename.replace('_0000', '')

        if modality == 'ct':
            shutil.copy(filepath, os.path.join(output_img_ct_dir, new_filename))

        elif modality == 'mr':
            shutil.copy(filepath, os.path.join(output_img_mr_dir, new_filename))

        else:
            print(f"ERROR : Unknown modality in file {filename} : {modality}")
```

```python
volume_dir = "../data/raw_pretraining/reorganized/imagesTr_topcow_ct"
segmentation_dir = "../data/raw_pretraining/reorganized/labelsTr_topcow_ct"

output_volumes_dir = "../data/processed_pretraining/imagesTr_topbrain_ct"
output_segmentations_dir = "../data/processed_pretraining/labelsTr_topbrain_ct"

preprocessing_pipeline = PreprocessingPipeline(volume_dir, segmentation_dir, output_volumes_dir, output_segmentations_dir, modality='CT', foreground_binarize=True)
preprocessing_pipeline.run()
```
